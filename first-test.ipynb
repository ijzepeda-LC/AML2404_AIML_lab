```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd


# Downloading imdb top 250 movie's data
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]




# create a empty list for storing
# movie information
list = []

# Iterating over movies to extract
# each movie's details
for index in range(0, len(movies[:3])):
    
    # Separating movie into: 'place',
    # 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"place": place,
            "movie_title": movie_title,
            "rating": ratings[index],
            "year": year,
            "star_cast": crew[index],
            }
    list.append(data)

# printing movie details with its rating.
for movie in list:
    print(movie['place'], '-', movie['movie_title'], '('+movie['year'] +
        ') -', 'Starring:', movie['star_cast'], movie['rating'])


##.......##
df = pd.DataFrame(list)
df.to_csv('imdb_top_250_movies.csv',index=False)
```

```python
url = "https://www.famousbirthdays.com/people/tom-cruise.html"
failed=[]
from datetime import datetime

def get_dob(actor):
    #replace odd chars
    actor=actor.lower().replace("'",'')
    actor=actor.replace(' ','-')
    #     if actor in not_working_list:
    #         return -1
    url = f"https://www.famousbirthdays.com/people/{actor}.html"
    print(url)
    # Send a GET request to the URL
    response = requests.get(url)

    # Create a BeautifulSoup object with the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    try:
        # Find the div element containing the person attributes
        attributes_div = soup.find("div", class_="bio-module__person-attributes")

        # Find the paragraph element containing the birthday information
        birthday_paragraph = attributes_div.find("p")

        # Find the span element with the birthday text
        birthday_span = birthday_paragraph.find("span", class_="type-16-18")

        # Extract the date of birth from the sibling span element
        dob = birthday_span.find_next_sibling("span").text.strip()

        print("Date of Birth:", dob)
        date_object = datetime.strptime(dob, "%B %d, %Y")
        year = date_object.year
    except:
        failed.append(actor)
        print(actor,"is not available - 404")
        year=1990
        
    return year
    
    
import time
toc=time.time()
data3=[]

# not_working_list

# for i,name in enumerate(set_actors):
for i,name in enumerate(list_actors):
    get_year=get_dob(name)
    #put get_yeat into df['year']
    #put name into df['name']
    actor={'name':name,'yob':get_year}
    data3.append(actor)
    if i%6==0:
        time.sleep(2)
    
df3=pd.DataFrame(data3)
df3.to_csv("actors.csv")
df3
```
